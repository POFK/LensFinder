````python
import astropy.io.fits as fits
import numpy as np
import os

import matplotlib.pylab as plt
````

````python
DirBase = "/data/inspur_disk03/userdir/wangcx"
````

````python
def show(data):
    fig, axes = plt.subplots(1,4,figsize=(4.5*4,4.5))
    ax = axes[0]
    ax.imshow(data[0])
    ax.invert_yaxis()
    ax = axes[1]
    ax.imshow(data[1])
    ax.invert_yaxis()
    ax = axes[2]
    ax.imshow(data[2])
    ax.invert_yaxis()
    ax = axes[3]
    ax.imshow(data.transpose(1,2,0))
    ax.invert_yaxis()
````

## Lens
### grade-A

````python
hdu=fits.open("/data/inspur_disk03/userdir/wangcx/decals_lens/BASS_resolution/grade-A/101t101/cutout_135.3125_9.9401.fits")
````

````python
data = hdu[0].data
data -= data.min()
data /= data.max()
````

````python
show(data)
````

````python
ldata = np.log10(data+1e-6)
ldata -= ldata.min()
ldata /= ldata.max()
````

````python
show(ldata)
````

````python
hdu.close()
````

### grade B

````python
with fits.open("/data/inspur_disk03/userdir/wangcx/decals_lens/BASS_resolution/grade-B/101t101/cutout_41.4742_-0.7052.fits") as hdu:
    data = hdu[0].data
    data -= data.min()
    data /= data.max()
````

### grade C

![](./check/cutout_16.2995_-0.0690.png)

````python
with fits.open("/data/inspur_disk03/userdir/wangcx/decals_lens/BASS_resolution/grade-C/101t101/cutout_16.2995_-0.0690.fits") as hdu:
    data = hdu[0].data
    data -= data.min()
    data /= data.max()
````
